```python
# Define some special matrices in the {e, f} subspace
I = Qobj([[0, 0, 0], [0, 1, 0], [0, 0, 1]])
sx = Qobj([[0, 0, 0], [0, 0, -1], [0, -1, 0]])
sy = Qobj([[0, 0, 0], [0, 0, -1j], [0, 1j, 0]])
sz = Qobj([[0, 0, 0], [0, 1, 0], [0, 0, -1]])

sx2 = np.matrix([[0, -1],[-1, 0]])
sy2 = np.matrix([[0, -1j],[1j, 0]])
sz2 = np.matrix([[1, 0],[0, -1]])
```

```python
# Function to simulate evolution of the general three-state system using the Lindblad Equation
def evolve_lindblad(J, γ_e, γ_f, Δ, tlist, psi0, expectations_needed=[sx, sy, sz]):
    
    # Define the basis vectors
    g = basis(3, 0)
    e = basis(3, 1)
    f = basis(3, 2)
    
    # The coupling Hamiltonian H_c
    H_tot = J*(tensor(e,f.dag()) + tensor(f,e.dag())) - (Δ / 2)*(tensor(f, f.dag())-tensor(e, e.dag()))
    H_tot.dims = [[3],[3]]
    
    # Add collapse operators L_e and L_f
    c_ops = []
    L_e = np.sqrt(γ_e) * tensor(g,e.dag())
    L_e.dims = [[3], [3]]
    L_f = np.sqrt(γ_f) * tensor(e,f.dag())
    L_f.dims = [[3], [3]]
    c_ops.append(L_e)
    c_ops.append(L_f)
    
    result = mesolve(H_tot, psi0, tlist, c_ops, expectations_needed)
    
    return result
```

```python
# Estimate the expectation of an operator empirically using post-selection
def single_empirical_expect(operator, rho, nshots=1000):
    normalized_p_ee = rho[1][0][1] / (rho[1][0][1] + rho[2][0][2])
    
    # p is the probability of being in the first excited state
    # op1 is the outcome if we are in the e state, op2 is the outcome if we are in the f state
    def trial(p, op1, op2):
        if random.random() < p:
            return op1
        else: return op2
    
    sum = 0
    outcome1 = expect(operator, basis(3, 1))
    outcome2 = expect(operator, basis(3, 2))
    for _ in range(nshots): sum += trial(normalized_p_ee, outcome1, outcome2)
    return sum / nshots    

def empirical_expect(operator, states):
    result = []
    
    for state in states: result.append(single_empirical_expect(operator, state))
    return result
```

```python
def post_selection_expect(operator, states):
    result = []
    
    for state in states: 
        result.append(expect(operator, state) / (state[1][0][1] + state[2][0][2]))
        result[-1] = result[-1].real
        
    return result
```

```python
def evolve_state(J, γ_e, γ_f, Δ, t, psi0):
    H = np.array([[Δ - 1j * γ_e / 2, J], [J, 0]])
    U = scipy.linalg.expm(-1j * H * t)
    
    return np.dot(U, psi0)

def evolve_list(J, γ_e, γ_f, Δ, tlist, psi0):
    results = []
    
    for t in tlist:
        results.append(evolve_state(J, γ_e, γ_f, Δ, t, psi0))
        
    return results

def expectation(operator, psi):
    return np.dot(psi.H, np.dot(operator, psi)).item((0, 0)).real

def norm_expectation(operator, psi):
    return np.dot(psi.H, np.dot(operator, psi)).item((0, 0)).real / np.dot(psi.H, psi).item((0, 0)).real

def all_expectations(operator, states):
    results = []
    
    for state in states: results.append(expectation(operator, state))
    return results

def all_norm_expectations(operator, states):
    results = []
    
    for state in states: results.append(norm_expectation(operator, state))
    return results

# evolved_state = np.dot(U, initial_state)
# print(np.dot(evolved_state.H, np.dot(sy2, evolved_state)))
#print(np.dot(evolved_state.H, np.dot(sy2, evolved_state)) / np.dot(evolved_state.H, evolved_state))
```

```python
def add_natural_damping(results, Gamma_0, final_time, avg=" "):
    if avg == " ": avg = np.average(results)
    
    npoints = len(results)
    for i in range(npoints):
        results[i] = avg + (results[i] - avg) * np.exp(-Gamma_0 * final_time * i / npoints)
```

```python
from scipy.signal import argrelextrema

def real_dlambda(J, γ_e, γ_f, Δ):
    H = np.array([[Δ - 1j * γ_e / 2, J], [J, 0]])
    
    lambdas, v = np.linalg.eig(H)
    return abs((lambdas[1] - lambdas[0]).real) / 2 / np.pi

def im_dlambda(J, γ_e, γ_f, Δ):
    H = np.array([[Δ - 1j * γ_e / 2, J], [J, 0]])
    
    lambdas, v = np.linalg.eig(H)
    return abs((lambdas[1] - lambdas[0]).imag)
```

```python
def estimate_period(J, γ_e, γ_f, Δ, t):
    tlist = np.linspace(0, t, t * 700)
    initial_state = np.matrix([[0], [1]])
    results2 = evolve_list(J, γ_e, γ_f, Δ, tlist, initial_state)
    
    sy_obs = np.array(all_norm_expectations(sy2, results2))
    maxima = argrelextrema(sy_obs, np.greater)
    
    return 1 / ((maxima[0][1] - maxima[0][0]) / 700)

def estimate_damping(J, γ_e, γ_f, Δ, t, Gamma_0 = 0.6):
    tlist = np.linspace(0, t, t * 700)
    initial_state = np.matrix([[0], [1]])
    results2 = evolve_list(J, γ_e, γ_f, Δ, tlist, initial_state)
    
    sy_obs = np.array(all_norm_expectations(sy2, results2))
    avg = np.average(sy_obs)
    sy_obs -= avg
    maxima = argrelextrema(sy_obs, np.greater)
    
    firstmax = maxima[0][0]
    secondmax = maxima[0][1]
    t_elapsed = (secondmax - firstmax) / 700
    return -np.log(sy_obs[secondmax] / sy_obs[firstmax]) / t_elapsed + Gamma_0
```

```python
γ_e = 7.1
γ_f = 0.14
J = 6.9
t = 2

Δ_vals = np.linspace(-8, 8, 50)

simulated_omega = []
simulated_gamma = []

for Δ in Δ_vals:
    simulated_omega.append(estimate_period(J, γ_e, γ_f, Δ * 2 * np.pi, t))
    simulated_gamma.append(estimate_damping(J, γ_e, γ_f, Δ * 2 * np.pi, t))
```

```python
Δ_vals = np.linspace(-8, 8, 200)
γ_e = 7.1
γ_f = 0.14
J = 6.9

reals = [real_dlambda(J, γ_e, γ_f, d * 2 * np.pi) for d in Δ_vals]

fig, axes = plt.subplots(1,1)
axes.set_xlabel(r'$\Delta / 2 \pi \;[MHz]$', fontsize=20)
axes.set_ylim(0, 8.5)
axes.plot(Δ_vals, reals, label="Theoretical")
axes.scatter(np.linspace(-8, 8, 50), simulated_omega, label="Actual")
axes.set_ylabel(r'$\Omega / 2 \pi \;[MHz]$', fontsize=15)
axes.legend()
```

```python
Δ_vals = np.linspace(-8, 8, 200)
γ_e = 7.1
γ_f = 0.14
J = 6.9
Gamma_0 = 0.6

reals = [im_dlambda(J, γ_e, γ_f, d * 2 * np.pi) + Gamma_0 for d in Δ_vals]

fig, axes = plt.subplots(1,1)
axes.set_xlabel(r'$\Delta / 2 \pi \;[MHz]$', fontsize=20)
axes.set_ylim([0, 5])
axes.plot(Δ_vals, reals, label="Theoretical")
axes.scatter(np.linspace(-8, 8, 50), simulated_gamma, label="Actual")
axes.set_ylabel(r'$\Omega / 2 \pi \;[MHz]$', fontsize=15)
axes.legend()
```

```python
def angle_to_state(theta, phi):
    return np.matrix([[np.cos(theta / 2)], [np.exp(1j * phi) * np.sin(theta / 2)]])

def calculate_deltaPfn(theta0, phi0, J, Δ, γ_e, γ_f, t):
    initial_state = angle_to_state(theta0, phi0)
    Pfn0 = float(abs(initial_state[1][0])) ** 2
    
    final_state = evolve_state(J, γ_e, γ_f, Δ, t, initial_state)
    Pfnfinal = float(abs(final_state[1][0])) ** 2 / np.dot(final_state.H, final_state).item((0, 0)).real

    return Pfn0 - Pfnfinal
```

```python
from scipy.signal import argrelextrema

def findeigenangles(deltaP, len_angle_range):
    to_return = []
    for J, J_series in enumerate(deltaP):
        Pabs = abs(J_series)
        maxima = np.array(argrelextrema(Pabs, np.less))
        if len(maxima[0]) >= 2: to_return.append(maxima[0] / len(J_series) * len_angle_range)
        
    return np.array(to_return)
```

```python
γ_e = 5.25
γ_f = 0.25
Δ = 0
t = 0.5 # Measured in microseconds = 500 ns

theta_range = np.linspace(np.pi, 0, 100)
J_range = np.linspace(0, 1.5, 100)

data = []
for J in J_range:
    data.append([])
    for theta in theta_range:
        data[-1].append(calculate_deltaPfn(theta, np.pi / 2, J, Δ, γ_e, γ_f, t))
             
data = np.array(data)

eigenangles = findeigenangles(data, np.pi)
inner_products = []
```

```python
for J, angles in enumerate(eigenangles):
    p1 = angle_to_state(angles[0], np.pi / 2)
    p2 = angle_to_state(angles[-1], np.pi / 2)
    inner_products.append(float(abs(np.dot(p1.H, p2))))

print(inner_products)
```
